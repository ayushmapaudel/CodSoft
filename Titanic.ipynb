```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("train.csv")

# Basic inspection
print(df.head())
print(df.info())
print(df.describe())

# Target variable distribution
print("Survival count:\n", df['Survived'].value_counts())
print("Mean Age:", df['Age'].mean())

# Survival rate by gender
print("Survival Rate by Sex:\n", df.groupby('Sex')['Survived'].mean())

# Check missing values
print("Missing values per column:\n", df.isnull().sum())

# Handle missing values
df = df.dropna(subset=['Age'])  # Drop rows where 'Age' is missing

# Optionally: fill missing 'Embarked' values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Handle categorical variable 'Sex'
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})

# Check for duplicates
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

# Final summary after cleaning
print(df.info())
```
